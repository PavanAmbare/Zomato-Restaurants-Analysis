# EXPLORATORY DATA ANALYSIS (EDA)

#  DATA CLEANING

```python
df.keys()
```

```python
# Deleting redundant columns.
```

```python
columns_to_drop=['url', 'address','phone','menu_item','reviews_list','cuisines']
df.drop(columns_to_drop,axis=1,inplace=True)
```

```python
df.head(1)
```

```python
# Renaming the columns.
```

```python
df.rename(columns={'name':'name_of_restro','book_table':'Table_Booking','rate':'rating','rest_type':'restro_type',
                  'approx_cost(for two people)':'approx_cost_for_two_people','listed_in(type)':'listed_type',
                   'listed_in(city)':'listed_in_place'},inplace=True)
df.head(1)
```

```python
# Cleaning individual columns.('rating')
```

```python
df['rating'].unique()
```

```python
df['rating']=df['rating'].replace(['NAN','NEW'],0)
```

```python
df['rating']=df['rating'].str.split('/').str[0]
```

```python
df['rating'].unique()
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating']=df['rating'].astype(float)
```

```python
df['rating'] = df['rating'].fillna(0)
```

```python
df['rating'].unique()
```

```python
# Cleaning individual columns.('name_of_restro')
```

```python
df['name_of_restro'].unique()
```

```python
df['name_of_restro']=df['name_of_restro'].str.split('-').str[0]
```

```python
df['name_of_restro']=df['name_of_restro'].str.lower()
```

```python
df['name_of_restro'].unique()
```

```python
# Cleaning individual columns.('location')
```

```python
df['location']=df['location'].str.split(',').str[0]
```

```python
df['location'].unique()
```

```python
df.isnull().sum()
```

```python
# Dropping dishes liked column
```

```python
df = df.drop('dish_liked',axis=1)
```

```python
# Remove the NaN values from the dataset
```

```python
most_frequent = df['restro_type'].mode()[0]
df['restro_type'] = df['restro_type'].fillna(most_frequent)
```

```python
df['approx_cost_for_two_people'] = pd.to_numeric(df['approx_cost_for_two_people'], errors='coerce')
```

```python
mean_cost= df['approx_cost_for_two_people'].mean()
df['approx_cost_for_two_people'].fillna(mean_cost,inplace=True)
```

#  Data Visualization

```python
#The percentage of the online orders in the restauants

df['online_order'].value_counts().plot(kind='pie', autopct="%0.1f%%", explode=[0, 0.1], shadow=True)
plt.title('Restaurants delivering Online or not')
plt.ylabel('Online Order Percentage')
plt.show()
```

```python
#The percentage of the booking a table in the restauants

colors = ['lightblue', 'lightcoral']
df['Table_Booking'].value_counts().plot(kind='pie', autopct="%0.1f%%", explode=[0, 0.2], shadow=True, colors=colors)

plt.title('Restaurants allowing table booking or not')
plt.ylabel('Book percentage')
plt.show()
```

```python
# Calculate average ratings
avg_ratings = df.groupby('Table_Booking')['rating'].mean().reset_index()

# Create a bar plot
sns.barplot(x='Table_Booking', y='rating', data=avg_ratings)

# Add labels and title
plt.xlabel('Table Booking')
plt.ylabel('Average Rating')
plt.title('Average Rating based on Table Booking Availability')

# Show plot
plt.show()
```
